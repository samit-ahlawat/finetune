```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\prog\cygwin\home\samit_000\nlp\nlpDocs\qa_fin2\FiQA_train_task2\final.csv")
df = df.loc[df.context.notna(), :].reset_index(drop=True)
df = df.loc[df.question.notna(), :].reset_index(drop=True)
df.head()
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df.loc[:, 'context'])
```

```python
best_n_clusters = 0
ss_dist = []
for n_cluster in range(4, 10, 1):
    kmeans = KMeans(n_clusters=n_cluster, random_state=42)
    kmeans.fit(X[0:1000, :])

    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    diff = X[0:1000, :] - centers[labels]
    ss = np.mean(np.einsum("ij,ij->i", diff, diff))
    ss_dist.append(ss) 

plt.plot(ss_dist)
```
